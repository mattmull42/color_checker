## 1. Opening the numpy archives

```python
import numpy as np
from os import path
```

```python
# CC archive
CC_archive_path = path.join('CC', 'CC.npz')
CC_dict = np.load(CC_archive_path)

print('CC archive :', CC_dict.files)

CC_cases_indices = CC_dict['color_checker_case']
CC_wavelengths = CC_dict['wavelengths']
CC_reflectances = CC_dict['reflectances']

print(CC_cases_indices.shape, CC_cases_indices) # The color checker cases
print(CC_wavelengths.shape, CC_wavelengths[:8]) # The 8 first wavelength samples
print(CC_reflectances.shape, CC_reflectances[0, :8]) # The reflectances of the first case and the 8 first wavelengths

CC_dict.close()
```

```python
# CC_GEOM archive
CC_GEOM_archive_path = path.join('CC_GEOM', 'CC_GEOM.npz')
CC_GEOM_dict = np.load(CC_GEOM_archive_path)

print('CC_GEOM archive :', CC_GEOM_dict.files)

CC_GEOM_cases_indices = CC_GEOM_dict['color_checker_case']
CC_GEOM_angles = CC_GEOM_dict['angles']
CC_GEOM_wavelengths = CC_GEOM_dict['wavelengths']
CC_GEOM_reflectances = CC_GEOM_dict['reflectances']

print(CC_GEOM_cases_indices.shape, CC_GEOM_cases_indices) # The color checker cases
print(CC_GEOM_angles.shape, CC_GEOM_angles) # The angle samples
print(CC_GEOM_wavelengths.shape, CC_GEOM_wavelengths[:8]) # The 8 first wavelength samples
print(CC_GEOM_reflectances.shape, CC_GEOM_reflectances[0, 0, :8]) # The reflectances of the first case, the first angle and the 8 first wavelengths

CC_GEOM_dict.close()
```

## 2. Ploting the reflectances

```python
from ploter import plot_reflectances
```

`plot_reflectance` parameters :
- `type` : (`str`) The type of data to plot. Can be either `CC`, `CC_GEOM` or `joint` (putting both `CC` and `CC_GEOM` on the same plot). Default is `joint`.
- `color_checker_case` : (`int`) The color checker's case, from 1 to 24 (see colorchecker-enumerated.png). If `type` is `CC`, `color_checker_case` can be set to `0` putting all the reflectances on the same plot.
- `save_name` : (`str` or `NoneType`) The name of the file in which the plot will be saved (in the file `plots/{type}/{save_name}.png`). Default is `None`, the plot will not be saved.

```python
plot_reflectances('CC', 0)
plot_reflectances('joint', 1)
```
